```python
import pandas as pd
filepath=('/content/drive/My Drive/titanic.csv')
data=pd.read_csv(filepath)
```

```python
df=pd.DataFrame(data)
```

```python
print("Number of rows and columns")
data.shape
```

```python
print("Column names and data types:")
print(df.dtypes)
```

```python
averageage = df['age'].mean()
print("Average age:", averageage)
```

```python
gender=df['sex'].value_counts
male_count=gender('male')
female_count=gender('female')
print(gender)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
gender_count=df['sex'].value_counts()
gender=gender_count.index
count=gender_count.values
plt.figure(figsize=(6,5))
plt.bar(gender,count,color=['b','g'])
plt.title('Number of Male and Female Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True,alpha=0.5)
plt.show()
```

```python
ages = df['age']
plt.figure(figsize=(8, 5))
plt.hist(ages, bins=20, color='b', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()









```

```python
survival_counts = df['survived'].value_counts()
labels = ['Not Survived', 'Survived']
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=90,colors=['g','red'])
plt.title('Survival Distribution')
plt.show()
```
