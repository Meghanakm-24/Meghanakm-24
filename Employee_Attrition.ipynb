#What is Co-relation ?

- Relationship between two  varibales and ranging from -1 to 1
- +1 : Perfect positive Corelation , If one variable increases , other variale increases proportionally
- -1 : Perfect Negative Corelation: If one variable is increasing , other variable decreses proportionally
- 0 : No  corelation: Two variables are not related to each other on linear manner
- Positive Co-relation: (greater than 0 upto +1)

- Example: Work hours- If employee is working for more hours--> evaluation points given by manager will increase (+ve corelation)

- Employye is working for more hours but he salary is low--> less satiesfaction (- ve corelation)
